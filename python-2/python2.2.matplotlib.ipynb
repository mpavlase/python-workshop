```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(10, 20)
print(x)
```

```python
count = x.size

data = np.random.rand(count)
data2 = np.random.rand(count)

plt.plot(x, data)
plt.plot(x, data2)
```

```python
# https://numpy.org/devdocs/user/absolute_beginners.html#plotting-arrays-with-matplotlib

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.15)
Y = np.arange(-5, 5, 0.15)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')
```

```python
x = np.linspace(0, np.pi * 2, 4)
x1 = np.linspace(0, np.pi * 2, 15)

plt.plot(x, np.sin(x))
plt.plot(x1, np.sin(x1))
```

```python
fig, axes = plt.subplots(2,3)
x = np.linspace(-1, 1)

for ax in axes.flat:
    ax.plot(x, x**2)
```
